### **A légköri nyomás magasságfüggését leírő rekurziós képlet származtatása** ###

**Felhasznált fizikai ismeretek, módszerek:**



1.   Az egyensúly feltétele, hogy a testre (a légkörelemre) ható erővektorok eredője nulla legyen $\sum \overrightarrow{F} = 0$.
2. $F = p \cdot A$ (Erő egyenlő nyomás szorozva felület), vagy másképpen $P=\frac{F}{A}$ (Nyomás egyenlő erő osztva felülettel).
3. A szétválaszthatóság elve szerint külön tudjuk vizsgálni a vízszintes irányú erőket és a függőleges irányú erőket. (A vízszintes irányú erők alakítják ki az atmoszféra nagy áramlási rendszereit, a ciklonokat és anti-ciklonokat, ezek az erők nem befolyásolják a függőleges irányú erőegyensúlyt, ezért ezekketl az erőkkel a továbbiakban nem foglalkozunk).
3. Fontos a problémához jól illeszkedő koordinátarendszer választása (x- és y-tengely vízszintes irányban, z-tengely a függőleges irány).
4. A számolás eredménye akkor elfogadható, ha a valóságra vonatkozó tapasztalatunk alátámasztja.

**Egy tetszőleges $z$ magasságban található, $\Delta z$ vastagságú légkörelemre ható erők**

Légkör_erőegyensúly.jpg

**Súlyerő**

A Föld tömegvonzásából származó súlyerő:

$F_z = m \cdot g = V \cdot \overline \rho \cdot g =A  \cdot \Delta z \cdot \overline \rho \cdot g$

ahol:


*   $m$: a légkörelem ösztömege,
*   $g$: a Föld tömegvonzásából származó nehézségi gyorsulás,
* $V$: a légkörelem térfogata,
* $\overline \rho$ : a légkörelem átlagos sűrűsége,
* $A$: a légkörelem alapterülete,
* $\Delta z$: a légkörelem magassága (vastagsága).







$\color{Red}{Kérdések: }$


1.   Miért kell átlagsűrűséggel számolni (a többi mennyiségnél miért nem használunk átlagolást)?
2.   Melyik az a $z$ magasság, ahol a sűrűség várhatóan a leginkább megegyezik a légkörelem átlagsűrűségével?





**Az átlagsűrűség becslése:**


$$\overline\rho  = \rho(z+\frac{\Delta z}{2}) $$

Ez alapján:

$$F_{súly} = A \cdot \Delta z \cdot \rho(z+\frac{\Delta z}{2}) \cdot g $$

Feltételeztük, hogy a sűrűség lineárisan (vagy közel lineárisan) változik a magassággal. (Ezt a feltételezést a későbbiekben ellenőrizni kell.)


$\color{Red}{Kérdés: }$ Hogyan írható fel az egyensúly?

**Az egyensúly egyenlete:**

$$\sum F = - A \cdot p(z+\Delta z) - A \cdot \Delta z \cdot \rho(z+\frac{\Delta z}{2}) \cdot g + A \cdot p(z) = 0$$

$\color{Red}{Feladat: }$ Értelmezzük az egyenletben szereplő tagok előjelét!

$\color{Red}{Kérdések: }$




1.   Milyen következtetést lehet levonni a nyomás magasság szerinti változására ebből az egyenletből?
2.   Hogyan lehet átrendezni az egyenletet?



**Átrendezés és egyszerűsítés**

$$ \frac{p(z+\Delta z)-p(z)}{\Delta z} = - \rho(z+\frac{\Delta z}{2}) \cdot g$$

$\color{Red}{Kérdés: }$ Mi ezzel az egyenlettel a probléma?

**Az ismeretlenek számának csökkentése**

$\color{Red}{Kérdés: }$ Melyik egyenlet teremt kapcsolatot a gáz nyomása és hőmérséklete között?

**A légkör állapotegyenlete**

$$\frac {p(z)}{\rho(z) \cdot T(z)} = \frac{R}{M}$$

Ahol:


1. $z$: a magasság, amelyre a légkör állapotegyenletét felírjuk
2. $\rho$: a légkörelem sűrűsége. $[\rho]=kg/m^3$
3.   $T$: a légkörelem hőmérséklete. $[T]= K$
4. $R$: univerzális gázállandó. $R=8,314 \frac{J}{mol \cdot K}$
5. $M$: a levegő moláris tömege. $M = 0.029 \frac{kg}{mol}$

$\color{Red}{Kérdés: }$ Jelen esetben milyen $z$ magasságra írjuk fel a légkör állapotegyenletét?



**A $(z+ \dfrac {\Delta z}{2})$ magasságra felírt állapotegyenlet behelyettesítése**

$$ \frac{p(z+\Delta z)-p(z)}{\Delta z} = - \frac{M \cdot g}{R \cdot T} \cdot p(z+\frac{\Delta z}{2})$$

$\color{Red}{Kérdések: }$

1.   Ebben az egyenletben a nyomás hány különböző magasságban szerepel?
2.   Hogyan lehet csökkenteni az egyenletben szereplő nyomások számát?



Lineáris nyomásfüggést feltételezve a légkörelemen belül:


$$ p(z+\frac{\Delta z}{2}) = \frac{p(z+\Delta z)+p(z)}{2}$$

$\color{Red}{Kérdés:} $ Miért feltételezhetünk lineáris nyomásfüggést a légkörelemen belül?


**Lineáris közelítés:** Kellően nagy felbontás mellett (pl. kellően rövid távolságokon) a legtöbb függvény közelíthető egy lineáris függéssel. Ezt a közelítést alkalmazzuk a jelen esetben is (a megoldás ismeretében a közelítés jogosságát igazolni kell).

A lineáris nyomásfüggéssel átírva az egyenletet:


 $$ \frac{p(z+\Delta z)-p(z)}{\Delta z} = - \frac{M \cdot g}{R \cdot T} \cdot \frac {p(z+\Delta z) +p(z)}{2}$$

Ez egy véges differencia egyenlet.


Átrendezés után:

$$p(z+\Delta z) = \frac{2 \cdot \frac{R \cdot T}{M \cdot g} -\Delta z}{2 \cdot \frac{R \cdot T}{M \cdot g} +\Delta z} \cdot p(z)$$



$\color{Red}{Kérdések:} $



1.   Mi az $\frac{R \cdot T}{M \cdot g}$ mennyiség dimenziója?
2.   Mekkorának célszerű választani $\Delta z$ értékét?
3. A rekurziós formula alkalmazásához kell egy referenciapont, ahol ismerjük a meghatározandó mennyiség értékét. Mi legyen ez a pont?




Válaszok:

1. $\big[\frac{R \cdot T}{M \cdot g} \big] = m$ (azaz hosszúság dimenziójú)

2. $\Delta z << \frac{R \cdot T}{M \cdot g}$
3. Referenciapontként (azaz a rekurzió kiinduló pontjaként) a tengerszinti magasságot választjuk: $p(z=0)=p_0 = 101 kPa$.




**Rekurzió szemléltetése**


Rekurzió szemléltetés.jpg

$\color{Red}{Kérdés: }$ Hogyan lehet kezelni azt a problémát, hogy a rekurziós egyenletben (a $H$ mennyiségben) szerepel a hőmérséklet?

Két lehetséges megoldás a légköri hőmérséklet ($T$) magasságfüggésének kezelésére:


1.   Közelítő megoldás: feltételezzük, hogy $T=áll.$ (Ebben az esetben az egyenlet analitikusan is megoldható.)
2.   A rekurzió alkalmazása során adott $z$ magasságra beírjuk a $T(z)$ tapasztalati értékét. A légkör legalsó rétegében a hőmérséklet $6,5\space ^ oC$-t csökken kilométerenként ($\frac{\Delta T}{\Delta z} = -6,5 \frac {\space ^oC}{km}$).



 $\color{Blue}{Feladat: }$ Írjunk programot a légköri nyomás és sűrűség magasságfüggésének számolására és ábrázolására! Ábrázoljuk (egy ábrán) a normalizált nyomást és sűrűséget is!

**Segédparaméterek**


1. A tengerszinti légnyomás: $p(z=0) = p_0 = 101 \space kPa$
2. Tengerszinti hőmérséklet: $T(z=0)=T_0 = 15 ^oC =288.15 K$
2. A hőmérséklet magasságfüggése:

$$ T(z) = T_0 -6.5 \frac{^oC}{km}\space \space \cdot z \space \space \space \space (0 < z < 11 \space km) $$


$$ T(z) = T_0 -6.5 \frac{^oC}{km}\space \space \cdot 11 \space km  \space \space \space \space (11 km < z) $$  

$\color{Blue}{Előkészítés :}$ Döntsük el, hogy Python lista vagy numpy tömb típusú változókat használunk a programban!

```python
import numpy as np
import matplotlib.pyplot as plt

R = 8.31447 # J/(mol x K)
M = 0.0289644 # kg/mol
g = 9.80665 # m/s^2
# Tengerszinti nyomás (Pa)
p0 = 101000

# Tengerszinti hőmérséklet (K)
T0 = 288.15

# A függőleges felbontás definiálása az első 10 km-re
z = np.linspace(0, 30000, 1000000)
Delta_z = z[1] - z[0]
p_z = np.zeros(len(z))
rho_z = np.zeros(len(z))
dT_dz = -6.5 # ^oC/km
# Hőmérséklet függvénye
def T(z):
    if z < 11000:
        Tz = T0  - dT_dz * z / 1000
    else:
        Tz = T0  - dT_dz * 11000 / 1000
    return Tz
# A H = RT/Mg függvény
def H(T):
    return R * T / (M * g)

# Kezdeti értékek
p_z[0] = p0

rho_z[0] = p_z[0] / (R * T(z[0]) / M)

for i in range(1, len(z)):
    p_z[i] = (2 * H(T(z[i])) - Delta_z) / (2 * H(T(z[i])) + Delta_z) * p_z[i - 1]
    rho_z[i] = p_z[i] / (R * T(z[i]) / M)

# Az iterációs részprogram meghívása
nyomás = p_z
sűrűség = rho_z
normalizált_nyomás = p_z/p_z[0]
normalizált_sűrűség = rho_z/rho_z[0]

plt.figure(figsize=(10, 5))

# Nyomás ábrázolása
plt.subplot(1, 2, 1)
plt.plot(z, nyomás)
plt.xlabel('Magasság (m)')
plt.ylabel('Nyomás (Pa)')
plt.title('Nyomás függvénye a magassággal')
plt.grid(True)


# Sűrűség ábrázolása
plt.subplot(1, 2, 2)
plt.plot(z, sűrűség)
plt.xlabel('Magasság (m)')
plt.ylabel('Sűrűség ($kg/m^3$)')
plt.title('Sűrűség függvénye a magassággal')
plt.grid(True)
plt.show(block=False)

# Normalizált nyomás és sűrűség ábrázolása
plt.subplot(2, 2, 1)
plt.plot(z, normalizált_nyomás)
plt.plot(z, normalizált_sűrűség)
plt.xlabel('Magasság (m)')
plt.ylabel('Normalizált érték')
plt.title('Normalizált érték függvénye a magassággal')
plt.legend(['Nyomás', 'Sűrűség'])
plt.grid(True)
plt.show(block=False)
```

```python
from scipy.integrate import solve_ivp

def atmospheric_model(z, y):
    p, rho = y
    Tz = T(z)
    H_T = H(Tz)
    dp_dz = -p / H_T
    drho_dz = -rho / H_T
    return [dp_dz, drho_dz]

rho0 = p0 / (R * T0 / M)

# Solve the system
z_span = (0, 30000)
z_eval = np.linspace(0, 30000, 100)
sol = solve_ivp(atmospheric_model, z_span, [p0, rho0], t_eval=z_eval, method='RK45')

# Extract results
z = sol.t
p_z, rho_z = sol.y

# Plot results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(z,p_z, label='Pressure (Pa)')
plt.xlabel('Magasság (m)')
plt.ylabel('Nyomás (Pa)')
#plt.legend()

plt.subplot(1, 2, 2)
plt.plot(z,rho_z, label='Density (kg/m³)', color='r')
plt.xlabel('Magasság (m)')
plt.ylabel('Sűrűség (kg/m³)')
#plt.legend()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(z,p_z/p_z[0], label='Pressure (Pa)')
plt.plot(z,rho_z/rho_z[0],"--", label='Density (kg/m³)', color='r')
plt.xlabel('Magasság (m)')
plt.ylabel('Érték')
plt.legend()
```

$\color{Blue}{Feladat:}$ Hasonlítsuk össze ezt a nyomásfüggést azzal az elképzelt esettel, amikor a hőmérséklet a légkörben állandó ($T = 288.15 \space K$)!

```python
import numpy as np
import matplotlib.pyplot as plt

R = 8.31447
M = 0.0289644
g = 9.80665
# Tengerszinti nyomás (Pa)
p0 = 101000

# Állandó hőmérséklet
T_állandó = 288.15

# A függőleges felbontás definiálása az első 10 km-re
z = np.linspace(0, 10000, 100)
p_z_közelített = np.zeros(len(z))
Delta_z = z[1] - z[0]

H_állandó = R * T_állandó / (M * g)
p_z_közelített[0] = p0
for i in range(1, len(z)):
    p_z_közelített[i] = (2 * H_állandó - Delta_z) / (2 * H_állandó + Delta_z) * p_z_közelített[i - 1]
nyomás_közelített = p_z_közelített
# Ábrázolás
plt.plot(z, nyomás)
plt.plot(z, nyomás_közelített)
plt.xlabel('Magasság (m)')
plt.ylabel('Nyomás (Pa)')
plt.title('Nyomás függvénye a magassággal')
plt.legend(['Változó hőmérséklet', 'Állandó hőmérséklet'])
plt.grid(True)
plt.show(block=False)
```

# **PFM_03_04 Függőleges hajítás (a tömegvonzás távolságfüggésének figyelembevételével)** #

## **Állandó gravitációs tér: analitikus megoldás** ##



$\color{Red}{Kérdés: }$ Milyen ismeretekre támaszkodva tudjuk a test ezt a problémát analitikusan megoldani?

**Felhasznált fizikai ismeretek, összefüggések**



1.   A $h=0$ magasságból (azaz a földfelszínről) $v_0$ sebességgel függőlegesen elindított test pályájának magassága és sebessége az alábbi képletek szerint változik az időben:

$$h(t) = v_0 \cdot t - \frac{1}{2} \cdot g \cdot t^2$$
$$v(t) = v_0 -g \cdot t$$
2. A test akkor éri el a pályája legmagasabb pontját, amikor $v_0 = g \cdot t$
3. Szimmetria okokból a test a $t = \frac{2 \cdot v_0}{g}$ időpillanatban ér vissza a felszínre.  


### $\color{blue}{Feladat: }$ ###
Egy testet függőlegesen indítunk a földfelszínről $100 km/óra $ kezdeti sebeséggel. Írjunk programot ami kiszámolja és ábrázolja a test pillanatnyi magasságát és a sebességét a fellővés utáni első 10 másodpercben!

$ \color {Blue}{Inicializálás:}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Gravitációs gyorsulás (m/s^2)
g = 9.81

# Kezdősebesség (m/s)
v0 = 100.0/3.6

# Időintervallum (másodperc)

# A szimuláció időtartalma és az időfelbontás definiálása (másodperc)
# t_values = np.linspace(0, 10, 100)
# A folyamat teljes ideje:
teljes_idő = 2 * v0/g
t_values = np.linspace(0,teljes_idő, 100)
```

$\color{Blue}{Számolás:}$

```python
# Aktuális sebesség függvénye
def sebesség(t):
  return v0 - g * t

# Aktuális magasság függvénye
def magasság(t):
  return v0 * t - 0.5 * g * t**2

# Sebesség és magasság kiszámítása az időintervallumon
sebességek = sebesség(t_values)
magasságok = magasság(t_values)
```

$\color{Blue}{Ábrázolás: }$

```python
plt.figure(figsize=(10, 5))

# Sebesség ábrázolása
plt.subplot(1, 2, 1)
plt.plot(t_values, sebességek,"o")
plt.xlabel('Idő (s)')
plt.ylabel('Sebesség (m/s)')
plt.title('Aktuális sebesség')
plt.grid(True)

# Magasság ábrázolása
plt.subplot(1, 2, 2)
plt.plot(t_values, magasságok,"o")
plt.xlabel('Idő (s)')
plt.ylabel('Magasság (m)')
plt.title('Aktuális magasság')
plt.grid(True)

plt.tight_layout()
plt.show()
```

 $\color{blue}{Feladat: }$ Az előző futtatás alapján alakítsuk át (ésszerűsítsük) a kódot!

## **Állandó gravitációs tér: numerikus megoldás (rekurzió)** ##



**Numerikus megoldás módszertana:**



1.   A $t=0$ időpillanatban ismerjük a test mozgásállapotát ($v=v_0$, $h=0$)
2.   A $t=0$ adatokból kiszámoljuk a sebességet és a magasságot a $t = \Delta t$ időpontra,
3. A $t=\Delta t$ adatokból kiszámoljuk a sebességet és a magasságot a  $t= 2 \cdot \Delta t$ időpontra,
4. Ezt az ún. rekurziót elvégezzük a teljes folyamatra.



### $\color{blue}{Feladat: }$ Számoljuk ki és ábrázoljuk a $v_0 = 100 \frac{m}{s}$ kezdősebeséggel függőlegesen fellőtt test magasságát és sebességét az idő függvényében és hasonlítsuk össze az eredményt az analítikus megoldással!###


```python
import numpy as np
import matplotlib.pyplot as plt

# Gravitációs gyorsulás
g0 = 9.81 # [g0]=m/s2

# Kezdősebesség
v0 = 100.0 # [v0] = m/s
teljes_idő = 2 * v0/g0
időfelbontás_numerikus = 100
t_numerikus = np.linspace(0,teljes_idő, időfelbontás_numerikus)
delta_t = t_numerikus[1] - t_numerikus[0]
sebesség_numerikus = np.zeros(len(t_numerikus))
magasság_numerikus = np.zeros(len(t_numerikus))
sebesség_numerikus[0] = v0
magasság_numerikus[0] = 0.0
for i in range (1, len(t_numerikus)):
  sebesség_numerikus[i] = sebesség_numerikus[i-1] - g0 * delta_t
  magasság_numerikus[i] = magasság_numerikus[i-1] + sebesség_numerikus[i-1] *delta_t  - 0.5 * g0 * delta_t**2

időfelbontás_analítikus = 100
t_analítikus = np.linspace(0,teljes_idő, időfelbontás_analítikus)
magasság_analítikus = v0 * t_analítikus - 0.5 * g0 * t_analítikus**2

# Magasság ábrázolása

plt.plot(t_numerikus, magasság_numerikus, "o", label = "Numerikus")
plt.plot(t_analítikus, magasság_analítikus, label ="Analítikus")

plt.xlabel('Idő (s)')
plt.ylabel('Magasság (km)')
plt.title('Aktuális magasság')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## **Változó gravitációs tér: numerikus megoldás (rekurzió)** ##


Megjegyzés: az állandó gravitációs térben való mozgással ellentétben a mostani probléma már tartalmaz közelítést.

**Az $R$ sugarú Föld felett $z$ magasságban található testre ható gravitációs erő:**

$$F(z)=G \cdot m \cdot M \frac{1}{{(R+z)}^2}$$

Ez alapján a Föld gravitációs terében mozgó test gyorsulása:

$$g(z)=G  \cdot M \frac{1}{{(R+z)}^2}$$

Jelöljük $g_0$-lal a nehézségi gyorsulást a földfelszínen ($z=0$):

$$g_0=G  \cdot M \frac{1}{R^2}$$

Ez alapján a képlet átírható:

$$g(z)=g_0 \left( \frac{R}{{R+z}} \right)^2 $$

$\color{Blue}{Feladat: }$ Ábrázoljuk a gravitációs gyorsulás magasságfüggését!



```python
import numpy as np
import matplotlib.pyplot as plt

G = 6.67430e-11 # Gravitációs állandó
M = 5.972e24 # A Föld tömege (kg)
R = 6371 * 1000 # A Föld sugara
g0 = G * M / (R * R)
maximális_magasság = float(input('Kérem a maximális magasságot (km): '))
maximális_magasság = maximális_magasság * 1000
z = np.linspace(0, maximális_magasság, 100)
g_z = G*M / (R + z)**2
plt.plot(z/1000, g_z)
plt.xlabel('Magasság (km)')
plt.ylabel('Gravitációs gyorsulás (m/s^2)')
plt.title('Gravitációs gyorsulás függvénye a magassággal')
plt.grid(True)
plt.show(block=False)
```

**A rekurziós képletek**

Jelöljük $\Delta t$-vel a $t_2$ és $t_1$ időpontok között eltelt időt: $t_2 = t_1 + \Delta t$.

1.   A $v_2$ sebesség felírható:

$$ v_2  = v_1 - g(z) \cdot \Delta t$$

2. A $z_2$ függőleges magasság felírható:

$$z_2 = z_1 + v_1 \cdot \Delta t - \frac{1}{2} \cdot g(z) \cdot \Delta t^2$$


```python
import numpy as np
import matplotlib.pyplot as plt

G = 6.67430e-11
# Föld tömege (kg)
M = 5.972e24
# Föld sugara
R = 6371 * 1000 # Föld sugara
g0 = G * M / (R * R)

def g(z):
    return g0 * (R / (R + z))**2

# Kezdősebesség (m/s)
v0 = 3132.0

időfelbontás = 0.1
teljes_idő = 2 * v0/g0
t_values = np.arange(0,teljes_idő, időfelbontás)
sebesség = np.zeros(len(t_values))
magasság = np.zeros(len(t_values))
sebesség_g0 = np.zeros(len(t_values))
magasság_g0 = np.zeros(len(t_values))
sebesség[0] = v0
sebesség_g0[0] = v0
magasság[0] = 0
magasság_g0[0] = 0
for i in range (1, len(t_values)):
    sebesség[i] = sebesség[i-1] - g(magasság[i-1]) * időfelbontás
    sebesség_g0[i] = sebesség_g0[i-1] - g0 * időfelbontás
    magasság_g0[i] = magasság_g0[i-1] + sebesség_g0[i-1] * időfelbontás + 0.5 * g0 * időfelbontás**2
    magasság[i] = magasság[i-1] + sebesség[i] * időfelbontás + 0.5 * g(magasság[i]) * időfelbontás**2
sebességek = sebesség
magasságok_km = magasság/1000
sebességek_g0 = sebesség_g0
magasságok_g0_km = magasság_g0/1000
maximális_magasság = np.max(magasságok_km)
print('Maximum magasság:', maximális_magasság, 'km')

# Sebesség ábrázolása
plt.subplot(1, 2, 1)
plt.plot(t_values, sebességek)
plt.plot(t_values, sebességek_g0)
plt.xlabel('Idő (s)')
plt.ylabel('Sebesség (m/s)')
plt.title('Aktuális sebesség')
plt.legend(['Változó g', 'Állandó g'])
plt.grid(True)

# Magasság ábrázolása
plt.subplot(1, 2, 2)
plt.plot(t_values, magasságok_km)
plt.plot(t_values, magasságok_g0_km)
plt.xlabel('Idő (s)')
plt.ylabel('Magasság (km)')
plt.title('Aktuális magasság')
plt.legend(['Változó g', 'Állandó g'])
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

R = 6371 * 1000
# Gravitációs gyorsulás a tengerszinten (m/s^2)
g0 = 9.81
def g(z):
  return g0 * (R / (R + z))**2

# Kezdősebesség (m/s)

időfelbontás = 1
v0 = 3132.0  # Replace with your initial velocity

# Initialize empty lists
sebesség = [v0]
sebesség_g0 = [v0]
magasság = [0]
magasság_g0 = [0]

i = 1
while True:
  if magasság[-1] < 0:  # Check if the last height value is negative
    break

  # Calculate new values and append to lists
  sebesség.append(sebesség[-1] - g(magasság[-1]) * időfelbontás)
  sebesség_g0.append(sebesség_g0[-1] - g0 * időfelbontás)
  magasság_g0.append(magasság_g0[-1] + sebesség_g0[-1] * időfelbontás + 0.5 * g0 * időfelbontás**2)
  magasság.append(magasság[-1] + sebesség[-1] * időfelbontás + 0.5 * g(magasság[-1]) * időfelbontás**2)

  i += 1
idő = range(len(magasság))
# Sebesség ábrázolása
plt.subplot(1, 2, 1)
plt.plot(idő, sebesség)
plt.plot(idő, sebesség_g0)
plt.xlabel('Idő (s)')
plt.ylabel('Sebesség (m/s)')
plt.title('Aktuális sebesség')
plt.legend(['Változó g', 'Állandó g'])
plt.grid(True)

# Magasság ábrázolása
plt.subplot(1, 2, 2)
plt.plot(idő, magasság)
plt.plot(idő, magasság_g0)
plt.xlabel('Idő (s)')
plt.ylabel('Magasság (km)')
plt.title('Aktuális magasság')
plt.legend(['Változó g', 'Állandó g'])
plt.grid(True)

plt.tight_layout()
plt.show()
```
