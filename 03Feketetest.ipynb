##**1.1. Alapfogalmak**##

**Spektrum**

A fényt alkotó elemi részecskék (fotonok) száma a fotonenergia függvényében.



*   Lehetséges $x$-tengelyek: fotonenergia, frekvencia és hullámszám ( a fotonenergiával arányos mennyiségek), hullámhossz (fordítottan arányos a fotonenergiával),   
*   Lehetséges $y$-tengelyek: energia, teljesítmény, intenzitás. (Megjegyzés: a későbbiekben elvégezzük az $y$-tengelyen ábrázolt mennyiségek értelmezésének pontosítását.)



**A fény duális természete**

Zavaró lehet a fény részecske-alapú illetve hullám-alapú leírása közötti ide-oda váltás. Sok esetben a hullám-alapú leírás (hullámhossz, hullámszám) teljesen elegendő. De pl. az eltérő hullámhosszú fény eltérő élettani hatásai miatt fontos, hogy ne feledkezzünk el a részecske alapú leírásról sem.  

**Hullámhossz ($\lambda$), hullámszám ($\nu$), frekvencia ($f$)**

Az alábbi mennyiségek a fény térbeli illetve időbeli periódikus ismétlődését reprezentálják:

*   Hullámhossz ($\lambda$): az a távolság, amekkora távolságonként a hullám ismétlődik ($\mu m=10^{-6}m$, $nm=10^{-9} m$).
*   Hullámszám ($\nu$): egységnyi hosszra eső ismétlődések száma $\mu = \dfrac 1\lambda$ ($1 \space 000 cm^{-1} \leftrightarrow 10 \mu m$, $10 \space 000 cm^{-1} \leftrightarrow 1 \mu m$).
* Frekvencia ($f$): egységnyi időre eső időbeli ismétlődések száma $f=c \cdot \nu$ ($c=3 \cdot 10^8 m/s$).

(A Nap esetében a hősugárzás infravörös része szinte teljes egészében a közeli-infravörös tartományba esik.)

**A hőmérsékleti sugárzás általános jellemzői**

Hőmozgás által keletkező energia elektromágneses sugárzás formájában történő kibocsátása.

A testek minden ($T > 0 K$) hőmérsékleten elektromágneses hullámokat bocsátanak ki, a környezet hőmérsékletétől függetlenül.

A sugárzás kibocsátásakor (emisszió) lényegében a test belső energiája átalakul elektromágneses energiává, a sugárzás elnyelésekor (abszorpció) pedig az elektromágneses energia alakul belső energiává.

A hőmérsékleti sugárzás emissziójával és abszorpciójával valósul meg a hőátadás egy lehetséges formája, a hősugárzás.

**A hőmérsékleti sugárzásos energiatranszport kvalitatív leírása**



* Az energiakibocsátás széles hullámhossztartományban történik (0< λ<
∞)
  
*   A sugárzás intenzitása és spektrális (hullámhossz vagy frekvencia szerinti) energiaeloszlása egy adott testnél csak a hőmérséklettől függ

*  A kibocsátás (emisszió) és az elnyelés (abszorpció) folyamata egymástól független.

* A testek akkor is sugároznak, ha a környezet miatt nem nyelnek el energiát.


**A hősugárzás és az energiaegyensúly**

Termodinamikai egyensúly esetén az emittált és az abszorbeált energia megegyezik, ellenkező esetben a sugárzó test melegszik vagy hűl a környezetéhez képest.

A két folyamat függetlensége azt jelenti, hogy a sugárzó test (nem-egyensúlyi állapotban) nem azt az energiát bocsátja ki, amit elnyelt.

**A hősugárzás (feketetestsugárzás) Plank-féle képlete**

$I(\lambda, T) = \frac{2 \cdot \pi \cdot h \cdot c^2}{\lambda^5} \cdot \frac{1}{\exp\left(\frac{h \cdot c}{\lambda \cdot k_B \cdot T}\right) - 1}$

ahol:
  
* $\lambda$: a hullámhossz ($[\lambda]=m$),
*   $T$: a test hőmérséklete ($[T]=K$),
*   $h = 6,6261\cdot 10^{-34}$:  Planck állandó ($[h]=J·s$),
*   $c = 3\cdot 10^8 \dfrac{m}{s}$:  fénysebesség,
*   $k_B = 1,381 \cdot 10^{-23} \dfrac{J}{K}$:  Boltzmann állandó.




$\color{Red}{Kérdés :}$ Mi az $I(\lambda,T)$ dimenziója és a dimenzió alapján mi a fizikai értelmezése?


**$I(\lambda,T)$ dimenziója és fizikai értelmezése**

$[I(\lambda,T)]=\dfrac{J}{s\cdot m^3}$

Egységnyi felület által egységnyi idő alatt, egységnyi szélességű hullámhossz-tartományban kibocsátott energia. (Egységnyi szélességű hullámhossz-tartomány által kibocsátott sugárzás intenzitása).

**Megjegyzések**



*   A $\pi$ szorzó a számlálóban onnan ered, hogy a teljes féltérre integráljuk a kibocsátott sugárzás energiáját. (A képlet $\pi$ szorzótényező-mentes változata egységnyi felület által egységnyi térszögben kisugárzott teljesítményre vonatkozik.)
*   Miután célszerű egyértelmű fizikai jelentéssel rendelkező  hullámhossz-tartomány-normálást alkalmazni,ezért az alábbiakban a számolt $I(\lambda,T)$ dimenziója $[I(\lambda,T)]=\dfrac{W}{m^2 \cdot nm}$ lesz. (Azaz $\Delta \lambda = 1 \space nm$ széles kibocsátási hullámhossztartományra fog vonatkozni.)



$\color{Red}{Kérdés :}$ Mi a hátránya illetve előnye ennek a mértékegységválasztásnak?

**A $\Delta \lambda = 1 \space nm$ választás következményei**

*   Hátrány: ez a választás  nem felel meg annak az általános elvnek, hogy  törekedni kell a mértékegységek egységesítésére.
*   Előny: $I(\lambda,T)$ ezzel a felírással jól értelmezhető.



**A $I(\lambda,T)$ görbén egy adott pont (azaz $I(\lambda_0,T_0)$) értelmezése**

A $T_0$ hőmérsékletű test által kibocsátott feketetestsugárzás integrális (össz) intenzitása a $\bigg(\lambda_0-\dfrac{\Delta \lambda}{02}\bigg) <\lambda  < \bigg(\lambda_0+\dfrac{\Delta \lambda}{2}\bigg)$ tartományban.

##**2.2. A Nap feketetestsugárzásának modellezése**##


$\color{Blue}{Feladat :}$ Írjunk kódot a Plank-féle sugárzási törvény ábrázolására a $T=5778 K$ esetre!

(Megjegyzés: a Föld globális energiaegyensúlyának szimulációjában  ezt a hőmérsékletértéket fogjuk használni  a Nap felszíni hőmérsékleteként).

````python
import numpy as np
import matplotlib.pyplot as plt

# Planck-törvény paraméterei
h = 6.62607015e-34  # Planck állandó (J·s)
c = 3.0e8           # Fénysebesség (m/s)
k_B = 1.380649e-23  # Boltzmann állandó (J/K)
delta_lambda = 1e-9 # A hullámhossztartomány szélessége = 1 nm

def planck_distribution(wavelength, temperature):
    term1 = (2 * np.pi * h * c**2) / wavelength**5
    term2 = 1 / (np.exp((h * c) / (wavelength * k_B * temperature)) - 1)
    return term1 * term2*delta_lambda

# Hullámhossz tartomány: 100 nm - 5000 nm, felbontás: 1 nm.
wavelengths = np.arange(100e-9, 5000e-9, 1e-9)  # méterben

# Hőmérséklet (K)
T_Sun_surface = 5778  # A Nap felszíni hőmérséklete

# Feketetest sugárzás számítása
intensity = planck_distribution(wavelengths, T_Sun_surface)

# Eredmények ábrázolása
plt.figure(figsize=(8, 6))
plt.plot(wavelengths * 1e9, intensity, label=f'T = {T_Sun_surface} K')
plt.title('Egységnyi felület, egységnyi térszög, egységnyi hullámhossztartomány')
plt.xlabel('Hullámhossz (nm)')
plt.ylabel('Intenzitás (W/m³/sr)')
plt.grid(True)
plt.legend()
plt.show()
````

$\color{Red}{Feladat :}$ Értelmezzük az eredményt!

**Napsugárzás intenzitásának mértéke**

Az ábra segítségével jól érzékelhető, hogy milyen hatalmas energiaforrásra (a Nap esetében a belsejében zajló nukleáris fúzióra) van szükség az ilyen mértékű energialeadás folyamatos biztosítására.

Gondoljunk arra, hogy a teljes kibocsátott feketetest-sugárzási teljesítmény egyenlő az integrális intenzitás és a Nap felszíni területének szorzatával!

$\color{Blue}{Kérdés :}$ Milyen adattípust érdemes használni az ábrázolás színeinek definiálásához?

````python
import numpy as np
import matplotlib.pyplot as plt

# Fizikai állandók
h = 6.62607015e-34  # Planck-állandó (J·s)
c = 3.0e8           # Fénysebesség (m/s)
k_B = 1.380649e-23  # Boltzmann-állandó (J/K)
delta_lambda = 1e-9
# Planck-törvény hullámhosszonkénti intenzitása
def planck_distribution(wavelength, temperature):
    term1 = (2 * h * np.pi*c**2) / wavelength**5
    term2 = 1 / (np.exp((h * c) / (wavelength * k_B * temperature)) - 1)
    return term1 * term2*delta_lambda

# Hullámhossz tartomány: 100 nm - 5000 nm, felbontás: 1 nm.
wavelengths = np.arange(100e-9, 5000e-9, 1e-9)  # méterben

T_Sun_surface = 5778  # Kelvin

# Feketetest sugárzás számítása
intensity = planck_distribution(wavelengths, T_Sun_surface)

# Hullámhossz-tartományok definiálása (nm-ben)
wl_ranges = {
    "UV-C": (100, 280),
    "UV-B": (280, 315),
    "UV-A": (315, 400),
    "Látható": (400, 700),
    "Infravörös": (700, 5000)
}

# Színek a spektrális tartományokhoz
colors = {
    "UV-C": "purple",
    "UV-B": "blue",
    "UV-A": "cyan",
    "Látható": "green",
    "Infravörös": "red"
}

# Ábrázolás
plt.figure(figsize=(10, 6))
plt.plot(wavelengths * 1e9, intensity, label=f'T = {T_Sun_surface} K', color='black')

# Hullámhossztartományok kitöltése
for ranges, (wl_min, wl_max) in wl_ranges.items():
  mask = (wavelengths * 1e9 >= wl_min) & (wavelengths * 1e9 <= wl_max)
  plt.fill_between(wavelengths[mask] * 1e9, intensity[mask], color=colors[ranges], alpha=0.3, label=ranges)

# Diagram beállításai
plt.title('Feketetest sugárzás spektrális tartományokkal')
plt.xlabel('Hullámhossz (nm)')
plt.ylabel('Intenzitás (W/m2/sr)')
plt.xlim(0,3000)
plt.grid(True)
plt.legend()
plt.show()
````

$\color{Blue}{Feladat :}$  Az előzőekben előállított intenzitáseloszlás-görbéből kiindulva  számoljuk ki a különböző spektrális tartományokba eső integrális intenzitást, illetve ezeknek az intenzitásoknak a százalékos arányát a teljes integrális intenzitáshoz képest!

*   *Energiaállapotok betöltöttsége*: az összenergia illetve a részecskeszám ismeretében optimalizálni tudjuk a modellezésben használt részecskék számát (pl. 99%).   





$\color{Blue}{Feladat :}$ Határozzuk meg $\lambda_{max,Föld}$ értékét az integrális módszerrel, úgy, hogy a modellezett hullámhossztartomány tartalmazza a teljes intenzitás 99%-át!

$\color{Blue}{Kérdés :}$ Milyen lépésekben (milyen művelettel) érdemes $\lambda_{max,Föld}$ értékét növelni?

````python
import numpy as np
import matplotlib.pyplot as plt

delta_lambda =  1e-9

# Spektrális teljesítmény integrálása az adott tartományokon belül
def integrate_spectrum(wavelengths, intensities, wavelength_range):
    mask = (wavelengths >= wavelength_range[0]) & (wavelengths <= wavelength_range[1])
    return np.trapz(intensities[mask], wavelengths[mask])/delta_lambda

sigma = 5.66e-8
# Hullámhossz tartomány: 100 nm - 100000 nm (referenciaértékhez)
wavelengths = np.arange(100e-9, 100e-6, 10e-9)

increase_factor = 1.2

# Hőmérsékletek: Nap és Föld
temperatures = {
    "Nap": 5778,    # Kelvin
    "Föld": 288     # Kelvin
}

results = {}

for name, T in temperatures.items():
    # Feketetest sugárzás számítása
    intensity = planck_distribution(wavelengths, T)

    # Tényleges integrális intenzitás egy nagy felső határral
    actual_total_intensity = sigma * temperatures[name]**4

    # Kezdeti feltételek
    infrared_limits = [5000]  # nm
    # Change here: specify wavelength range as [lower_limit, upper_limit]
    integrated_intensities = [integrate_spectrum(wavelengths, intensity, [100e-9, 5000e-9])]
    # Iteratív növelés, amíg el nem érjük a 99%-ot
    while integrated_intensities[-1] < 0.99 * actual_total_intensity:
        new_limit = infrared_limits[-1] * increase_factor
        if new_limit > 100000:
            break

        new_integral = integrate_spectrum(wavelengths, intensity, [100e-9, new_limit * 1e-9])

        infrared_limits.append(new_limit)
        integrated_intensities.append(new_integral)

    results[name] = (infrared_limits, integrated_intensities, actual_total_intensity)

# Eredmények ábrázolása százalékos formában
plt.figure(figsize=(8, 5))

for name, (infrared_limits, integrated_intensities, actual_total_intensity) in results.items():
    percent_values = [integral / actual_total_intensity * 100 for integral in integrated_intensities]
    plt.plot(infrared_limits, percent_values, marker='o', linestyle='-', label=name)

plt.axhline(y=99, color='r', linestyle='--', label="99% határ")
plt.xlabel('Infravörös határ (nm)')
plt.ylabel('Integrált intenzitás (% a teljeshez képest)')
plt.title('Integrált spektrális intenzitás növekedése a hullámhosszhatárral')
plt.legend()
plt.grid()
plt.show()
````

A Wien-féle eltolódási törvény kapcsolatot teremt a $T$ hőmérséklet és a $\lambda_{max}$ között: $\lambda_{max} \cdot T = áll.$

$\lambda_{max}$: az a hullámhossz ahol maximális a kibocsátott intenzitás.

**Hullámhosszeltolódás**

##**3.2. A Wien törvény konstans tagjának meghatározása**##



$\color{Blue}{Feladat:}$ Több hőmérsékleten elvégzett számolással határozzuk meg a konstans értékét a Wien féle eltolódási törvényben!

````python
import numpy as np
import matplotlib.pyplot as plt

# Planck-törvény hullámhosszonkénti intenzitása
def planck_distribution(wavelength, temperature):
    term1 = (2 * h *np.pi *c**2) / wavelength**5
    term2 = 1 / (np.exp((h * c) / (wavelength * k_B * temperature)) - 1)
    return term1 * term2*delta_lambda

# Hőmérsékletek definiálása
temperatures = [100, 200, 500, 1000, 2000, 5000, 10000]  # Kelvin

# Hullámhossztartomány a csúcs kereséséhez
wavelengths = np.linspace(1e-9, 300e-6, 100000)  # Hullámhossz (1 nm - 300 µm)

# Csúcs hullámhosszak kiszámítása
lambda_max_list = []
for T in temperatures:
    intensities = planck_distribution(wavelengths, T)*1e-9
    lambda_max = wavelengths[np.argmax(intensities)]  # Maximális intenzitás helye
    lambda_max_list.append(lambda_max)

# Wien-állandó numerikus meghatározása
lambda_max_array = np.array(lambda_max_list)
T_array = np.array(temperatures)
b_computed = np.mean(lambda_max_array * T_array)

# Wien-állandó elméleti értéke
b_theoretical = 2.897771955e-3  # m·K

# Ábrázolás
plt.figure(figsize=(10, 6))
plt.scatter(T_array, lambda_max_array * 1e9, color='red', label='Numerikus eredmények ($\lambda_{max}$)')
plt.plot(
    T_array,
    b_theoretical / T_array * 1e9,
    label=f'Elméleti: $b / T$, $b = {b_theoretical:.3e}$ m·K',
    color='blue',
    linestyle='--'
)
plt.title('Wien-féle eltolódási törvény igazolása')
plt.xlabel('Hőmérséklet (K)')
plt.ylabel('Csúcs hullámhossz ($\lambda_{max}$, nm)')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.show()

# Wien-állandó eltérés
error_percentage = abs(b_computed - b_theoretical) / b_theoretical * 100

# Eredmények kiíratása
print(f"Numerikusan számolt Wien-állandó: {b_computed:.6e} m·K")
print(f"Elméleti Wien-állandó: {b_theoretical:.6e} m·K")
print(f"Hiba: {error_percentage:.6f} %")
````

#**4. A Stefan-Boltzmann törvény**#

$\color{Blue}{Feladat:}$ Igazoljuk, hogy a kibocsátott összintenzitás a hőmérséklet negyedik hatványával arányos!

````python
import numpy as np

# Fizikai állandók
h = 6.62607015e-34  # Planck-állandó (J·s)
c = 3.0e8           # Fénysebesség (m/s)
k_B = 1.380649e-23  # Boltzmann-állandó (J/K)

# Stefan-Boltzmann állandó elméleti értéke
sigma_theoretical = (2 * np.pi**5 * k_B**4) / (15 * h**3 * c**2)

# Planck-törvény integrálása dimenzió nélküli változóval
def planck_integral_temperature(T):
    """
    Integráljuk a Planck-törvényt dimenzió nélküli változóval.
    T: Hőmérséklet (K)
    """
    x = np.linspace(1e-6, 50, 100000)  # Dimenzió nélküli változó
    integrand = (x**3) / (np.exp(x) - 1)  # Dimenzió nélküli integrandus
    integral = np.trapz(integrand, x)  # Trapéz módszer
    factor = (2 * np.pi * k_B**4 * T**4) / (h**3 * c**2)  # Tényező
    return factor * integral

# Hőmérsékletek definiálása
temperatures = [100, 200, 500, 1000, 2000, 5000, 10000]  # Kelvin

# Integrált teljesítmények számítása
results = [planck_integral_temperature(T) for T in temperatures]

# Stefan-Boltzmann törvény illesztése
T_array = np.array(temperatures)
P_array = np.array(results)
fit_coeffs = np.polyfit(np.log(T_array), np.log(P_array), 1)

# Kiszámított Stefan-Boltzmann állandó
sigma_computed = np.exp(fit_coeffs[1])

# Hibaszámítás
error_percentage = abs(sigma_computed - sigma_theoretical) / sigma_theoretical * 100

sigma_computed, sigma_theoretical, error_percentage
````
